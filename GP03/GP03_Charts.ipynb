### Visualizing The Gender Gap In College Degrees

##### Starter code

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt

women_degrees = pd.read_csv('GP03-percent-bachelors-degrees-women-usa.csv')
cb_dark_blue = (0/255,107/255,164/255)
cb_orange = (255/255, 128/255, 14/255)
stem_cats = ['Engineering', 'Computer Science', 'Psychology', 'Biology', 'Physical Sciences', 'Math and Statistics']

fig = plt.figure(figsize=(18, 3))

for sp in range(0,6):
    ax = fig.add_subplot(1,6,sp+1)
    ax.plot(women_degrees['Year'], women_degrees[stem_cats[sp]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[stem_cats[sp]], c=cb_orange, label='Men', linewidth=3)
    ax.spines["right"].set_visible(False)    
    ax.spines["left"].set_visible(False)
    ax.spines["top"].set_visible(False)    
    ax.spines["bottom"].set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_title(stem_cats[sp])
    ax.tick_params(bottom="off", top="off", left="off", right="off")
    
    if sp == 0:
        ax.text(2005, 87, 'Men')
        ax.text(2002, 8, 'Women')
    elif sp == 5:
        ax.text(2005, 62, 'Men')
        ax.text(2001, 35, 'Women')
plt.show()
```

#### Generating 6 by 3 plot
17 degrees total, three groups of 6, 5, and 6

```python
# Degree groups
stem_cats = ['Psychology', 'Biology', 'Math and Statistics', 'Physical Sciences', 'Computer Science', 'Engineering']
lib_arts_cats = ['Foreign Languages', 'English', 'Communications and Journalism', 'Art and Performance', 'Social Sciences and History']
other_cats = ['Health Professions', 'Public Administration', 'Education', 'Agriculture','Business', 'Architecture']

fig = plt.figure(figsize=(15, 18))

#Text labels locations dictionary
text_loc = {
    'Psychology':{
        'Mx':2005,'My':85,
        'Wx':2002,'Wy':10
    },
    'Engineering':{
        'Mx':2005,'My':89,
        'Wx':2002,'Wy':6
    },
    'Foreign Languages':{
        'Mx':2005,'My':77,
        'Wx':2002,'Wy':18
    },
    'Health Professions':{
        'Mx':2005,'My':92,
        'Wx':2002,'Wy':3
    },
    'Architecture':{
        'Mx':2005,'My':65,
        'Wx':2002,'Wy':30
    }
}

# Group plotting function
# Parameters: list of group_titles, plotting figure, column # to plot in
def plot_group(group_titles, fig, col):
    for sp in range(0,len(group_titles)):
        ax = fig.add_subplot(6,3,3*sp+col+1)
        ax.plot(women_degrees['Year'], women_degrees[group_titles[sp]], c=cb_dark_blue, label='Women', linewidth=3)
        ax.plot(women_degrees['Year'], 100-women_degrees[group_titles[sp]], c=cb_orange, label='Men', linewidth=3)
        ax.spines["right"].set_visible(False)    
        ax.spines["left"].set_visible(False)
        ax.spines["top"].set_visible(False)    
        ax.spines["bottom"].set_visible(False)
        ax.set_xlim(1968, 2011)
        ax.set_ylim(0,100)
        ax.set_yticks([0,100])
        ax.set_ylim(0,100)
        ax.axhline(50, c=(171/255, 171/255, 171/255), alpha=0.3)
        ax.set_title(group_titles[sp])
        ax.tick_params(bottom=False, top=False, left=False, right=False, labelbottom=False)

        # Place text labels
        if group_titles[sp] in text_loc:
            tl = text_loc[group_titles[sp]]
            ax.text(tl['Mx'], tl['My'], 'Men')
            ax.text(tl['Wx'], tl['Wy'], 'Women')
            
        # Remove ticks for all charts except the bottom ones
        if sp == len(group_titles)-1:
            ax.tick_params(labelbottom=True)

degree_groups = [stem_cats, lib_arts_cats, other_cats]
# plot three groups of degrees
for i in range (0,3):
    plot_group(degree_groups[i], fig, i)
    
plt.savefig('GP03_plot.png')
plt.show()
    
```
